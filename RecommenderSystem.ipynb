```python
retail_data.dropna(inplace=True)
```

```python
retail_data['CustomerID'] = retail_data['CustomerID'].apply(lambda x: 'C'+str(int(x)))
retail_data.head()
```

```python
# Dataframe for building the recommendation system
orders = retail_data[['CustomerID', 'Description', 'InvoiceNo', 'Quantity' ]]
orders.head()
```

```python
orders = orders[
    orders['Description'].str.upper() == orders['Description']
]
orders.head()
```

```python
# top ten products (non-personalized)
orders.groupby('Description').agg(orders=('InvoiceNo', 'nunique')).sort_values(by='orders', ascending=False).head(10)
```

```python
# products that were sold the most
orders.groupby('Description').agg(quantity=('Quantity', 'sum')).sort_values(by='quantity', ascending=False).head(10)
```

```python
# generating the user item interaction matrix
int_matrix = orders.pivot_table(index="CustomerID", columns=['Description'], values='Quantity').fillna(0)
int_matrix.head()
```

```python
active_user = 'C12346'
```

```python
recommendations = int_matrix.corrwith(int_matrix.loc[active_user], axis=1)
recommendations.dropna(inplace=True)
recommendations
```
